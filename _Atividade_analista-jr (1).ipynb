```python
#importa a função tree.DecisionTreeRegressor da biblioteca scikit-learn
from sklearn.tree import DecisionTreeRegressor
#importa a função model_selection.train_test_split da biblioteca scikit-learn
from sklearn.model_selection import train_test_split
#importa a função metrics.mean_absolute_error da biblioteca scikit-learn
from sklearn.metrics import mean_absolute_error
```

```python
#separa os onjuntos de dados nir e lab em dados de treino e dados de validação
treino_nir, vali_nir, treino_lab, vali_lab = train_test_split(nir, lab, random_state=1)

#treino: dados de treino
#vali: dados de validação
#random_state=1: Controla o embaralhamento aplicado aos dados antes de aplicar a divisão
```

```python
#Função para testar os dados de treino e verificar a media de erro do modelo
def get_valide(max_leaf_nodes, treino_x, vali_x, treino_y, vali_y):
    #armazena na veriavel teste a função de treino
    teste = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    #max_leaf_nodes: define a melhor quantidade de nós para a servire ser mais acertiva
    #random_state: Controla a aleatoriedade do estimador

    #treina o teste com base nas colunas x e y de treino
    teste.fit(treino_x, treino_y)
    #.fit: atributo usado treinar os dados com base nos dados (x,Y)

    #realiza a previsão com base nos dados de validação
    vali_previsoes = teste.predict(vali_x)
    #.predict atributo usado prever os dados de x com base em uma nova tabela y

    #calcula a media de erro que o foi obtido na previsão com os dados de validação e armazena na variavel erro_media
    erro_medio = mean_absolute_error(vali_y, vali_previsoes)
    
    #reotna a varival erro_medio
    return(erro_medio)
```

```python
#aramazena uma margem de erro alta na variavel erro
erro = 1e10

for nos in range(2, 1001):
    #range conta de 2 até 1000 para definir o melhor valor para max_leaf_nodes

    # aramazena o erro medio a ser rentornado na função get_valide na variavel erro_nos
    erro_nos = get_valide(nos, treino_nir, vali_nir, treino_lab, vali_lab)
    
    #se o erro gerado fa função for menor que a margem definida
    if erro_nos < erro:
        #a margem de erro passa a ser o erro calculado plea função
        erro = erro_nos
        #o menor erro calculado passa aser o melhor erro
        melhor_erro = nos
    
    #escreve na tela quantos nós foram analizados
    print("Nós analaizados:", nos, "nós")
```

```python
#realiza o teste de margem de erro com o melhor valor retornado no loop e calcula a margem de erro
teste_error = get_valide(melhor_erro, treino_nir, vali_nir, treino_lab, vali_lab)

#retorna da tela o melhor erro e qual é a margem calculada
print("Melhor erro geral foi com", melhor_erro, "Nós", "- Erro médio:", teste_error)
```

```python
#armazena na variavel modeloFinal o melhor modelo de treino
modeloFinal = DecisionTreeRegressor(max_leaf_nodes=melhor_erro, random_state=1)

#treino o novo modelo com todos os dados
modeloFinal.fit(nir,lab)
```
